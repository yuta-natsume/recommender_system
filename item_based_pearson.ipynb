```python
def find_best_corr(rating_a, rating_b):
#     print(rating_a, rating_b)
    corr_ab = np.corrcoef(rating_a, rating_b)
#     print(corr_ab)
    arranged_corr = 1 - corr_ab
#     print(arranged_corr)
#     print(arranged_corr[0, 1])
    return arranged_corr[0, 1]
```

```python
def create_model(rating_matrix, k=20):
    """
    :param R : numpy array of item representations
    :param k : number of nearest neighbors to return    
    :return model : our knn model
    """    
    model = NearestNeighbors(metric='correlation', n_neighbors=k+1, algorithm='brute')
    model.fit(rating_matrix)    
    return model
```

```python
def nearest_neighbors(rating_matrix, model):
    """
    compute the top n similar items for each item.    
    :param rating_matrix : items representations
    :param model : nearest neighbors model    
    :return similarities, neighbors
    """    
    similarities, neighbors = model.kneighbors(rating_matrix)    
#     display(similarities)
#     display(neighbors)
    return similarities, neighbors
```

```python
def predict(userid, itemid, df_similarities, df_neighbors):
    """
    Make rating prediction for user userid on item itemid    
    :param userid : id of the active user
    :param itemid : id of the item for which we are making prediction        
    :return r_hat : predicted rating
    """
    dim = len(df_similarities.columns)
    # Get items similar to item itemid with their corresponding similarities
    item_neighbors = df_neighbors.loc[itemid, :]
    item_similarities = df_similarities.loc[itemid, :]
    
    # get ratings of user with id userid
    uratings = ratings_200[userid]
    
    # similar items rated by item the user of i
    scores = ratings_200.loc[item_neighbors, userid].to_numpy().reshape(1, dim)
    
    sims = item_similarities.to_numpy().reshape(dim, 1)
    
    dot = np.dot(scores, sims)
    som = np.sum(np.abs(sims))

    if som == 0:
        return np.nan
    elif dot == 0:
        return 0
    
    return dot / som
```

```python
def calculate_corr(df_score, ratings_200, idx):
    start_list = [2500, 2500, 3500, 4500, 5500, 6500]
    end_list = [9509, 3499, 4499, 5499, 6499, 9509]
    corr_list = []
    ratings_200 = ratings_200.drop('movie_id', axis=1)
    for i in range(len(start_list)):
        divided_output = df_score.loc[start_list[i]:end_list[i], :]
        true_output = ratings_200.loc[start_list[i]:end_list[i], :]
        
        nan_idx = divided_output.loc[np.isnan(divided_output[idx]), idx].index
#         print(nan_idx)
        output_nonan = divided_output.drop(nan_idx)
#         display(output_nonan)
        true_nonan = true_output.drop(nan_idx)
#         display(true_nonan)
        
        corr = output_nonan[idx].corr(true_nonan[idx])
        corr_list.append(corr)
        
    return corr_list
```

```python
def make_df_corr(corr_list, df_corr, idx):
    df_corr.loc[idx, :] = corr_list
    return df_corr
```
